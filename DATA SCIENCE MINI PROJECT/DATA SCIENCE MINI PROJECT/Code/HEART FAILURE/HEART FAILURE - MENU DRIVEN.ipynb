```python
## import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")

df=pd.read_csv("heart.csv")

ch=0

while ch < 8:
    print("\n",'\033[1m'+'DATA SCIENCE MINI PROJECT on HEART FAILURE'+'\033[0m')
    print('\033[1m' + '1' + '\033[0m', "Accessing the Data")
    print('\033[1m' + '2' + '\033[0m', "How to check for duplicate Data?")
    print('\033[1m' + '3' + '\033[0m', "How to check for Null values?")
    print('\033[1m' + '4' + '\033[0m', "Describe")
    print('\033[1m' + '5' + '\033[0m', "What is Distribution?")
    print('\033[1m' + '6' + '\033[0m', "Categorical representation")
    print('\033[1m' + '7' + '\033[0m', "3D Plotting")
    print('\033[1m' + '0' + '\033[0m', "EXIT")
    
    ch=int(input("Enter your choice :  "))
    
    if ch == 1:
        print(df)
        print(df.info())
        print(df.shape)
        
    if ch == 2:
        temp=df.duplicated().sum()
        print("Total number of duplicate values : ",temp)
    
    if ch == 3:
        temp=df.isnull().sum()
        print("Total number of null values :")
        print(temp)
    
    if ch == 4:
        temp=df.describe()
        print(temp)
    
    if ch ==5:
        
        ch1=0
        
        while ch1 < 4:
            print('\033[1m' + '1' + '\033[0m', "Distribution with respect to", '\033[1m' + 'Age' + '\033[0m')
            print('\033[1m' + '2' + '\033[0m', "Distribution with respect to", '\033[1m' + 'Sex' + '\033[0m')
            print('\033[1m' + '3' + '\033[0m', "Distribution with respect to", '\033[1m' + 'ExerciseAngina' + '\033[0m')
            print('\033[1m' + '0' + '\033[0m', "EXIT")
            
            ch1=int(input("Enter your choice: \n"))
            
            if ch1 == 1:
                #Age is normaly distributed
                plt.figure(figsize=(16,8))
                sns.displot(df,x="Age",bins=5)#New diplot function used instead of distplot
                plt.show()
                plt.figure(figsize=(16,8))
                sns.displot(df,x="Age",hue="Sex",bins=25)
                plt.show()
            
            if ch1 == 2:
                #Aproxx 79% is male and 21 % is female
                plt.figure(figsize=(16,6))
                plt.subplot(1,2,1)
                sns.countplot(df['Sex'])
                plt.subplot(1,2,2)
                df['Sex'].value_counts().plot(kind='pie',autopct="%.2f")
                plt.show()
                
            if ch1 == 3:
                plt.figure(figsize=(16,6))
                plt.subplot(1,2,1)
                sns.countplot(df['ExerciseAngina'])
                plt.subplot(1,2,2)
                df['ExerciseAngina'].value_counts().plot(kind='pie',autopct="%.2f")
                plt.show()
            
            if ch1 == 0:
                break
            
    if ch == 6:
        
        ch1=0
        
        while ch1 <3:
            
            print('\033[1m' + '1' + '\033[0m', "Categorical with respect to Sex")
            print('\033[1m' + '2' + '\033[0m', "Categorical with respect to ChestPainType")
            print('\033[1m' + '0' + '\033[0m', "EXIT")
                  
            ch1=int(input("Enter your choice:\n"))
            
            if ch1==1:
                x=px.bar(df,'HeartDisease','Age',color='Sex')
                x.show()
                fig = px.box(df,"HeartDisease","Age", color="Sex")
                fig.update_traces(marker_size=10)
                fig.show()
            
            if ch1==2:
                fig = px.box(df,"HeartDisease","Age", color="ChestPainType")
                fig.update_traces(marker_size=10)
                fig.show()
                
            if ch1==0:
                break    
            
    if ch == 7:
        #3D plot with Age ,RestingBP,MaxHR
        fig = px.scatter_3d(df, x='Age', y='RestingBP', z='MaxHR',color='ChestPainType',symbol='HeartDisease')
        fig.show()
        #3D plot with Age ,RestingBP, Cholesterol,
        fig = px.scatter_3d(df, x='Age', y='RestingBP', z='Cholesterol',color='ChestPainType',symbol='HeartDisease')
        fig.show()
        
    if ch == 0:
        print("",'\033[1m'+'THANK YOU'+'\033[0m')
        break
```
