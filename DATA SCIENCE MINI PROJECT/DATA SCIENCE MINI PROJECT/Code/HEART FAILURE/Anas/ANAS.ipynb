```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
df=pd.read_csv("heart.csv")
ch=0
while ch < 7:
    print("        ",'\033[1m'+'DATA SCIENCE MINI PROJECT'+'\033[0m')
    print("        ",'\033[1m'+'HEART FAILURE'+'\033[0m')
    print("1.Accessing the",'\033[1m' + 'DATA' + '\033[0m',"also shows the sahpe and info")
    print("2.How to check for",'\033[1m' + 'DUPLICATE DATA' + '\033[0m')
    print("3.How to check for",'\033[1m' + 'NULL VALUES' + '\033[0m')
    print("4.",'\033[1m' + 'DESCRIBE' + '\033[0m'," the data")
    print("5.what is the ",'\033[1m' + 'DISTRIBUTION' + '\033[0m')
    print("6.",'\033[1m' + 'CATAGORICAL' + '\033[0m',"representaion")
    print("7.",'\033[1m' + '3-D Ploting' + '\033[0m')
    print("0.To Exit")
    ch=int(input("Enter your chpice :  "))
    if ch == 0:
        break
    if ch == 1:
        print(df)
        print(df.info())
        print(df.shape)
    if ch == 2:
        temp=df.duplicated().sum()
        print("Total number of duplicate values : ",temp)
    if ch == 3:
        temp=df.isnull().sum()
        print("Total number of null values :")
        print(temp)
    if ch == 4:
        temp=df.describe()
        print(temp)
    if ch ==5:
        ch1=0
        while ch1 < 4:
            print("0.To go back")
            print("1.Distribution with respect to",'\033[1m' + 'Age' + '\033[0m')
            print("2.Distribution with respect to",'\033[1m' + 'Sex' + '\033[0m')
            print("3.Distribution with respect to",'\033[1m' + 'ExerciseAngina' + '\033[0m')
            
            ch1=int(input("Enter your choice: \n"))
            if ch1 == 0:
                break
            if ch1 == 1:
                #Age is normaly distributed
                plt.figure(figsize=(16,8))
                sns.displot(df,x="Age",bins=5)#New diplot function used instead of distplot
                plt.show()
                plt.figure(figsize=(16,8))
                sns.displot(df,x="Age",hue="Sex",bins=25)
                plt.show()
            if ch1 == 2:
                #Aproxx 79% is male and 21 % is female
                plt.figure(figsize=(16,6))
                plt.subplot(1,2,1)
                sns.countplot(df['Sex'])
                plt.subplot(1,2,2)
                df['Sex'].value_counts().plot(kind='pie',autopct="%.2f")
                plt.show()
            if ch1 == 3:
                plt.figure(figsize=(16,6))
                plt.subplot(1,2,1)
                sns.countplot(df['ExerciseAngina'])
                plt.subplot(1,2,2)
                df['ExerciseAngina'].value_counts().plot(kind='pie',autopct="%.2f")
                plt.show()
    if ch == 6:
        ch1=0
        while ch1 <3:
            print("1.Catagorical with respect to",'\033[1m' + 'Sex' + '\033[0m')
            print("2.Catagorical with respect to",'\033[1m' + 'ChestPainType' + '\033[0m')
            ch1=int(input("enter yout choice:\n"))
            if ch1==0:
                break
            if ch1==1:
                x=px.bar(df,'HeartDisease','Age',color='Sex')
                x.show()
                fig = px.box(df,"HeartDisease","Age", color="Sex")
                fig.update_traces(marker_size=10)
                fig.show()
            if ch1==2:
                fig = px.box(df,"HeartDisease","Age", color="ChestPainType")
                fig.update_traces(marker_size=10)
                fig.show()
            
    if ch == 7:
        #3D plot with Age ,RestingBP,MaxHR
        fig = px.scatter_3d(df, x='Age', y='RestingBP', z='MaxHR',color='ChestPainType',symbol='HeartDisease')
        fig.show()
        #3D plot with Age ,RestingBP, Cholesterol,
        fig = px.scatter_3d(df, x='Age', y='RestingBP', z='Cholesterol',color='ChestPainType',symbol='HeartDisease')
        fig.show()
```
